```python
na_data = train_df.isnull().sum()/len(train_df)
na_data = na_data[na_data != 0.0]
na_data.sort_values(ascending=False)
```

```python
na_data = test_df.isnull().sum()/len(test_df)
na_data = na_data[na_data != 0.0]
na_data.sort_values(ascending=False)
```

```python
for col in ("PoolQC", "MiscFeature", "Alley", "Fence", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "BsmtFinType2", "BsmtExposure", "BsmtFinType1", "BsmtCond", "BsmtQual", "MasVnrType"):
    train_df[col] = train_df[col].fillna("None")
    test_df[col] = test_df[col].fillna("None")
```

```python
train_df["LotFrontage"] = train_df.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
test_df["LotFrontage"] = test_df.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
for col in ("GarageYrBlt", "GarageArea", "GarageCars", "BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF", "BsmtFullBath", "BsmtHalfBath", "MasVnrArea"):
    train_df[col] = train_df[col].fillna(0.0)
    test_df[col] = test_df[col].fillna(0.0)
```

```python
for col in ("Electrical", "MSZoning", "Utilities", "SaleType", "KitchenQual", "Exterior2nd", "Exterior1st"):
    train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
    test_df[col] = test_df[col].fillna(train_df[col].mode()[0])
```

```python
test_df["Functional"] = test_df["Functional"].fillna("Typ")
```

```python
na_data = train_df.isnull().sum()/len(train_df)
na_data = na_data[na_data != 0.0]
na_data.sort_values(ascending=False)
```

```python
na_data = test_df.isnull().sum()/len(test_df)
na_data = na_data[na_data != 0.0]
na_data.sort_values(ascending=False)
```

```python
train_df.to_csv("../data/interpolated_train.csv", index=False)
test_df.to_csv("../data/interpolated_test.csv", index=False)
```

```python
for col in ("MSSubClass", "OverallCond", "YrSold", "MoSold", "OverallQual"):
    train_df[col] = train_df[col].astype(str)
    test_df[col] = train_df[col].astype(str)
```

```python
train_df["TotalSF"] = train_df["TotalBsmtSF"] + train_df["1stFlrSF"] + train_df["2ndFlrSF"]
test_df["TotalSF"] = test_df["TotalBsmtSF"] + test_df["1stFlrSF"] + test_df["2ndFlrSF"]
train_df["TotalBath"] = train_df["BsmtFullBath"] + 0.5*train_df["BsmtHalfBath"] + train_df["FullBath"] + 0.5*train_df["HalfBath"]
test_df["TotalBath"] = test_df["BsmtFullBath"] + 0.5*test_df["BsmtHalfBath"] + test_df["FullBath"] + 0.5*test_df["HalfBath"]
```
